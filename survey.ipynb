```python
#replacing missing values as 'nan' and changing categorical variables to type 'category'
df = df.replace(' ',np.nan)
for column, series in df.iteritems():
    if not column in ['Name', 'Prison_ID', 'ID']:
        df[column] = df[column].astype('category')
```

```python
#seperate dataframe for attitudinal scales
columns_to_keep = ['Prison', 'Name', 'Prison_ID', 'ID', 'Gender', 'Group', 'Data_Round', 'Samples', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17',
       'Q18', 'Q19', 'Q20', 'Q21', 'Q22', 'Q23', 'Q24', 'Q25', 'Q26','Q27', 'Q28', 'PostQ30']
attitudinal_scales_df = pd.DataFrame(df,columns = columns_to_keep)
attitudinal_scales_df.tail(5)
```

```python
for column, series in df.iteritems():
    if column in ['Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17','Q18', 'Q19', 'Q20', 'Q21',
                  'Q22', 'Q23', 'Q24', 'Q25', 'Q26','Q27', 'Q28', 'PostQ30']:
        attitudinal_scales_df[column] = attitudinal_scales_df[column].astype('string')
```

```python
attitudinal_scales_df = attitudinal_scales_df.replace('Strongly Agree',1)
attitudinal_scales_df = attitudinal_scales_df.replace('Somewhat Agree',2)
attitudinal_scales_df = attitudinal_scales_df.replace('Not Sure',3)
attitudinal_scales_df = attitudinal_scales_df.replace('Somewhat Disagree',4)
attitudinal_scales_df = attitudinal_scales_df.replace('Strongly Disagree',5)
```
